```python
import pandas as pd

# Ścieżka do pliku CSV
file_path = 'pogodynka_zbilansowana.csv'

# Wczytaj dane do DataFrame
df = pd.read_csv(file_path)

# Sprawdź pierwsze wiersze
df.head()
```

```python
# Informacje ogólne
df.info()

# Statystyki opisowe
df.describe()

# Sprawdzenie braków danych
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram temperatury
plt.figure(figsize=(8,5))
sns.histplot(df['TemperaturaPowietrza'], kde=True)
plt.title('Rozkład temperatury')
plt.show()
```
